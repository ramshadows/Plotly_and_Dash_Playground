```python
Iris_versicolor =  iris.loc[iris["class"] == "Iris-versicolor"]["petal_length"]
```

```python
Iris_virginica =  iris.loc[iris["class"] == "Iris-virginica"]["petal_length"]
```

```python
hist_data = [Iris_setosa, Iris_versicolor, Iris_virginica]
```

```python
group_labels = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
fig = ff.create_distplot(hist_data=hist_data, group_labels=group_labels)
```

```python
pyo.plot(fig, filename="distplot6.html")
```
